```python
import pandas as pd
```

```python
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df
```

```python
df.drop(['customerID'], inplace = True, axis = 1)
```

```python
df.dtypes
```

```python
for i in ['Partner', 'Dependents', 'PhoneService', 'MultipleLines','OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']:
    df[i]=df[i].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
df = pd.concat([df, dummy1], axis = 1)
```

```python
df = df.drop(['InternetService'], axis =1)
```

```python
df.Contract.unique()
```

```python
dummy2 = pd.get_dummies(df['Contract']);
```

```python
df = pd.concat([df, dummy2], axis = 1)
```

```python
df = df.drop(['Contract'], axis =1)
```

```python
dummy3 = pd.get_dummies(df['PaymentMethod']);
```

```python
df = pd.concat([df, dummy3], axis = 1)
```

```python
df = df.drop(['PaymentMethod'], axis =1)
```

```python
df.Churn=df.Churn.apply(lambda x: 1 if x=="Yes" else 0)
```

```python
df['TotalCharges'].unique()
```

```python
df = df[df["TotalCharges"].str.strip() != ""]
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scalingcolums = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[scalingcolums] = scaler.fit_transform(df[scalingcolums])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size = 0.2)
```
